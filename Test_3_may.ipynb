```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("sales.csv",encoding = "unicode_escape")
df
```

```python
df.head(5)
```

```python
r=df.drop("unnamed1",axis=1)
r
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

# 3.Drop NULL value

```python
s=df.dropna(axis=1)
s
```

# 4.What is the average age of customers?

```python
df.pivot_table(index=['User_ID','Gender'],values=['Age'],aggfunc='mean')
```

# 5.Who are the majority buyers?(Genderwise) 

```python
p=df.query("Orders>3 and Gender=='F'")[["Gender","Orders","Cust_name"]]
p
```

# 6.Where do our majority customers fall?(Age)

```python
x=df["Age Group"]
plt.hist(x,bins=20)
plt.title("Majority Customers Age")
plt.xlabel("Age Group")
plt.show()
```

# 7.Which state has highest sales?

```python
plt.figure(figsize=(15,5))
sns.set_style("darkgrid")

sns.countplot(x=df["State"],palette="Set3",edgecolor="black")

plt.title("State with highest sales",size=30)
plt.xlabel("State",size=15)
plt.ylabel("Amount",size=15)
plt.yticks(size=10)
plt.xticks(size=10)
plt.show()
```

# 8.How does occupation effects the purchase?

```python
plt.figure(figsize=(15,5))

sns.countplot(x=df["Occupation"],palette="Set3",edgecolor="black")

plt.title("Effect of Occupation on Purchase",size=30)
plt.xlabel("Occupation",size=15)
plt.ylabel("Purchase",size=15)
plt.show()
```
